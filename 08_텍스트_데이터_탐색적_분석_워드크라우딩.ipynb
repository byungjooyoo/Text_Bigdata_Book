## 8.2.3 웹 크롤링에 의한 데이터 수집과 저장

```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
#Keyword 검색과 news_items 확보
keyword ='인공지능'
max_page = 2
news = ""
for p in range(max_page):
    url = f"https://search.hankookilbo.com/Search?Page={p+1}&tab=NEWS&sort=relation&searchText={keyword}&searchTypeSet=TITLE,CONTENTS&selectedPeriod=%EC%A0%84%EC%B2%B4&filter=head"
    resp = requests.get(url)
    html_src = resp.text
    soup = bs(html_src, 'html.parser')
    news_items = soup.select('div.img-box')
    # 각 뉴스 링크를 생성하고 기사를 종합
    for item in news_items:
        link = item.find('a').get("href")
        resp = requests.get(link)
        html_src = resp.text
        soup = bs(html_src, 'html.parser')
        # 기사에서 패러그래프 종합
        text=''
        for para in soup.select('p.editor-p'):
            text += para.text.strip()
        news +=text
        news +="\n\n"
news
```

```python
with open("/content/drive/MyDrive/Data/인공지능.txt", 'w') as f:
    f.write(news)
```

## 8.3.1 데이터 필터링

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/byungjooyoo/Dataset/main/movie_comment.csv')
df = df[['Title', 'Score','Comment']]
df
```

```python
con = (df.Title == '인질') & (df.Score == 10)
df.loc[con, ['Score', 'Comment']]
```

```python
con1 = df['Title']=='인질'
df1=df.loc[con1, : ].tail(10)
df1.drop(8121, axis=0, inplace= True)
df1.drop("Title", axis = 1, inplace =True)
con2 = df1['Score'] == 1
con3 = df1['Score'] == 6
df1.loc[con2 | con3, : ]
```

## 8.3.2 결측치 및 중복 데이터 제거

```python
print(len(df), df.isnull().sum())
df.dropna(inplace = True)
len(df)
```

## 8.3.4 하나의 문서로 변환

```python
data1 = data.loc[data['Label'] == 1, ['Comment']]
data0 = data.loc[data['Label'] == 0, ['Comment']]
[x for x in data1['Comment']]
```

```python
text1 = ''
for para in data1['Comment']:
  text1 += para+'. '
print(text1)

text0 = []
for para in data0['Comment']:
  text0.append(para)
text0 = ".".join(text0)
print(text0)
```

# 08.4 탐색적 빈도분석

## 8.4.1 형태소 분석기를 이용한 명사 추출

```python
!pip install konlpy
from konlpy.tag import Komoran
komoran = Komoran()
nouns1 = komoran.nouns(text1)
nouns0 = komoran.nouns(text0)
nouns0[:10]
```

## 8.4.2 데이터 정제와 빈도 추출

```python
nouns1 = [n for n in nouns1 if len(n) > 1]
nouns0 = [n for n in nouns0 if len(n) > 1]
```

```python
from collections import Counter
count1 = Counter(nouns1)
count0 = Counter(nouns0)
top1 = count1.most_common(100)
top0 = count0.most_common(100)
print(top1)
print(top0)
```

```python
!wget https://raw.githubusercontent.com/byungjooyoo/Dataset/main/korean_stopwords.txt
import re
stop_words=[]
with open("korean_stopwords.txt", 'r') as f:
    while True:
        line = f.readline().strip()
        stop_words.append(line)
        if not line: break
print(stop_words)
```

```python
stop_words = ['황정민', '영화']+stop_words
dic1 = dict(top1)
dic0 = dict(top0)
for word in stop_words:
    if word in dic1:
        del dic1[word]
    if word in dic0:
        del dic0[word]
print(dic1)
print(dic0)
```

## 8.4.3 빈도 그래프

```python
import nltk
import matplotlib.pyplot as plt
pos = nltk.Text(nouns1)
neg = nltk.Text(nouns0)
plt.rc('font', family='NanumGothic')
plt.figure(figsize=(8,8))
pos.plot(30)
```

# 08.5 분석결과 시각화

## 8.5.1 워드클라우딩 단순 사례

```python
from wordcloud import WordCloud
from konlpy.tag import Komoran
import matplotlib.pyplot as plt
from konlpy.corpus import kolaw
# 대상 텍스트 데이터
kolaw.fileids()
para = kolaw.open('constitution.txt').read()
tagger = Komoran()
nouns = tagger.nouns(para)
text = ", ". join(nouns)
# 클라우딩 객체 설정
my_path='/content/NanumGothic.ttf'
wordcloud = WordCloud(font_path=my_path,
                      background_color='white',
                      width=1200,
                      height=800).generate(text)
fig = plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```
